```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
import numpy as np
```

### Captação dos dados

```python
df = pd.read_csv('https://gist.githubusercontent.com/omadson/770934cb0df9e2d2c3e8db031f1139a5/raw/4d181efc035b327d9d66256915951384517c603e/arrecadacao.csv')
df.head()
```

### Normalização dos dados

```python
#Modifica um das categorias do dado
df =df.assign(arrecadacao_milhao = df.arrecadacao / 1000000)
```

### Ordena as barras

```python
df = df.sort_values(by = 'arrecadacao', ascending=False)
```

### Altera a disposição das linhas

```python
sns.set_style("whitegrid")
```

### Plotando 

```python
plt.figure(figsize = (5,2))
ax = sns.barplot(data = df, x = 'arrecadacao_milhao', y = 'titulo', color = "red")

#Colocando os respectivos valores em cada barra
ax.bar_label(
    ax.containers[0],
    fmt = '%.1f',
    label_type = 'center',
    color = 'white'
)

#Adicionando legendas
plt.xlabel('Arrecadação (Milhões de dólares)') #Titulo na abcissa
plt.ylabel('Filme') #Titulo na ordenada
plt.title('Arrecadação de filmes')

#Linha vertical na origem e retirando as bordas
ax.set(frame_on=False)
ax.axvline(0, color="k", clip_on=False)
plt.show()
```
