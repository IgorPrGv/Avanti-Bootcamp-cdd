## Informações estatísticas (tendência central e dispersão)

### Passageiros do titanic

#### >Variáveis qualitativas

```python
vars_quali = ["survived", "pclass", "sex", "embarked", "class", "who", "adult_male", "deck", 
               "embark_town", "alive", "alone"]

moda = tnt[vars_quali].mode()

print('Moda')
display(moda)
```

#### >Variáveis quantitativas

```python
vars_quanti = ["age", "sibsp", "parch", "fare"]

media = tnt[vars_quanti].mean()
moda = tnt[vars_quanti].mode()
mediana = tnt[vars_quanti].min()
desvio_padrao = tnt[vars_quanti].std()

print('Média')
display(media)
print()
print('Moda')
display(moda)
print()
print('Mediana')
display(mediana)
print()
print('Desvio padrão')
display(desvio_padrao)
```

## Estatísticas para cada variável

### Passageiros do titanic

```python
vars_quali = ["survived", "pclass", "sex", "embarked", "class", "who", "adult_male", "deck", 
               "embark_town", "alive", "alone"]

#Loop para percorrer todas as variáveis
for variavel in vars_quali:
    #Recebe cada variavel
    unique_value = tnt[variavel].unique()
    for value in unique_value:
        print(variavel, ' - ', value)
        #Recolhe, por lógica booleana, a condiçao especifica da vez no loop interno da variável em análise 
        subamostra = tnt[tnt[variavel] == value]
        display(subamostra.describe())
        print()
```

### Arrecadação de gorjetas

```python
vars_quali = ["sex", "smoker", "day", "time"]

#Loop para percorrer todas as variáveis
for variavel in vars_quali:
    #Recebe cada variavel
    unique_value = tps[variavel].unique()
    for value in unique_value:
        print(variavel, ' - ', value)
        #Recolhe, por lógica booleana, a condiçao especifica da vez no loop interno da variável em análise 
        subamostra = tps[tps[variavel] == value]
        display(subamostra.describe())
        print()
```
